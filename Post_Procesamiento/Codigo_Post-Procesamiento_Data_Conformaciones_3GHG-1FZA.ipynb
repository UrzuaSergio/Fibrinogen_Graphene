De este nuevo archivo se obtendra los valores de energias de Solvatación y Coulomb para cada conformación, con el objetivo de calcular la Energia Libre de Interacción. Que puede ser definida como:

$\nabla \nabla G_{int} = (E_{total})_{Interaccion} - \sum_{k=1}^{N_{comp}} (E_{total_{k}})_{Aisladas}$

$E_{total} = E_{solv} + E_{coul} + E_{surf}$

Tambien se requieren los valores de energias para los sistemas por separados. Estos pueden obtenerse de los archivos _output_ repectivos. Se debera crear un archivo que contenga los valores de energia libre de interacción para cada conformación. 

```python
import numpy
import sys
import os


################################################################################################################
#Archivos output proteina superficie interactuando
################################################################################################################

Output_interaccion = '3GHG_Graph_Results/3GHG-Interaccion_Phi-File_0-180'
file_aux= open('auxiliar.txt','w')
file_1 = open(Output_interaccion,'r')

E_solv_interaccion = []
h = []
a_tilt = []
a_rot = []
E_coul_interaccion = -349371501.174910

for line_1 in file_1:
    linea_1=line_1.split()
    if len(linea_1)==11:
        if linea_1[0]== '2.00':
            aa='{0} {1} {2} {3} \n'.format(linea_1[0],linea_1[2],linea_1[4],linea_1[6])
            file_aux.write(aa)
            
            
file_aux.close()

file_aux= open('auxiliar.txt','r')
X=numpy.loadtxt(file_aux)
Xc=numpy.copy(X)
E_solv_interaccion[:]=Xc[:,3]
h[:]=Xc[:,0]
a_tilt[:]=Xc[:,1]
a_rot[:]=Xc[:,2]
file_aux.close()


##############################################################################################################
#Funcion para leer archivo_output de PyGBe
##############################################################################################################

def scanOutput(filename):
    
    flag = 0 
    files = []
    file1=open(filename,'r')
    for line in file1:
        line = line.split()
        if len(line)>0:
            if line[0]=='Converged':
                iterations = int(line[2])
            if line[0]=='Total' and line[1]=='elements':
                N = int(line[-1])
            if line[0]=='Totals:':
                flag = 1 
            if line[0]=='Esolv' and flag==1:
                Esolv = float(line[2])
            if line[0]=='Esurf' and flag==1:
                Esurf = float(line[2])
            if line[0]=='Ecoul' and flag==1:
                Ecoul = float(line[2])
            if line[0]=='Time' and flag==1:
                Time = float(line[2])
            if line[0]=='Reading':
               files.append(line[-1])  

    return N, iterations, Esolv, Esurf, Ecoul, Time, files

##############################################################################################################
##############################################################################################################
#Proteina aislada
##############################################################################################################
N_prot_aislada = []
iterations_prot_aislada = []
Esolv_prot_aislada = []
Esurf_prot_aislada = []
Ecoul_prot_aislada = []
Time_prot_aislada = []

N_run, iterations_run, Esolv_run, Esurf_run, Ecoul_run, Time_run, files =scanOutput('3GHG_Graph_Results/3GHG_Aislada_out')

N_prot_aislada.append(N_run)
iterations_prot_aislada.append(iterations_run)
Esolv_prot_aislada.append(Esolv_run)
Esurf_prot_aislada.append(Esurf_run)
Ecoul_prot_aislada.append(Ecoul_run)
Time_prot_aislada.append(Time_run)

#####################################################################################################################
#####################################################################################################################
#Superficie Bidimensional de Grafeno
#####################################################################################################################
N_graph_aislada = []
iterations_graph_aislada = []
Esolv_graph_aislada = []
Esurf_graph_aislada = []
Ecoul_graph_aislada = []
Time_graph_aislada = []

N_run1, iterations_run1, Esolv_run1, Esurf_run1, Ecoul_run1, Time_run1, files1 =scanOutput('3GHG_Graph_Results/surf2d_Aislada_out')

N_graph_aislada.append(N_run1)
iterations_graph_aislada.append(iterations_run1)
Esolv_graph_aislada.append(Esolv_run1)
Esurf_graph_aislada.append(Esurf_run1)
Ecoul_graph_aislada.append(Ecoul_run1)
Time_graph_aislada.append(Time_run1)


#####################################################################################################################
#####################################################################################################################
#Calculo energia de interaccion y encontrar estado de minima Energia Libre de Interacción.

E_interaccion = numpy.zeros(len(h))

for i in range(len(h)):
    E_interaccion[i]= (E_solv_interaccion[i] + E_coul_interaccion) - (Esolv_prot_aislada[0] + Ecoul_prot_aislada[0] + Esolv_graph_aislada[0] + Ecoul_graph_aislada[0])

E_mas_prob=min(E_interaccion)
argumento_mas_prob=numpy.argmin(E_interaccion)
h_mas_prob = h[argumento_mas_prob]
a_tilt_mas_prob=a_tilt[argumento_mas_prob]
a_rot_mas_prob=a_rot[argumento_mas_prob]

print('Distancia proteina sensor =',h_mas_prob, ' Angstrom')
print('Alpha_tilt =', a_tilt_mas_prob, '°')
print('Alpha_rot =', a_rot_mas_prob, '°')
print('Energia de Interaccion =', E_mas_prob, ' [Kcal/mol]')

#Archivo_input_probability_plot_python
conformacion_file = open('3GHG_Graph_Results/E_int_3GHG-GRAPH_Conformacion_0-180_v2','w')
for i in range(len(h)):
    angulos_energia_interaccion='{0} {1} {2} \n'.format(a_tilt[i], a_rot[i], E_interaccion[i])
    conformacion_file.write(angulos_energia_interaccion)
conformacion_file.close()

os.system('rm -r auxiliar.txt')
os.system('rm -r auxiliar_2.txt')	

print('Se han generados archivos requeridos para graficar distribución de Probabilidades')
```

```python
from numpy import *
import matplotlib as mpl
import matplotlib.pyplot as plt

def rotation_integration(U, tilt, rot):

    i=0
    da_rot = 0
    while da_rot==0:
        da_rot = rot[i+1]-rot[i]
        i+=1
    print('da_rot = %f'%(da_rot))

    nrot = int(360/da_rot)
    point_prob = []
    point_tilt = []
    point_rot  = []

    kT = 2.479/4.184
    rot_integral = []
    angles = []
    da_tilt = []
    for i in range(len(tilt)):
        # Extreme cases: tilt = 0 and tilt = 180
        if abs(tilt[i])<1e-10:
            rot_integral.append(exp(-U[i]/kT)*2*pi)
            angles.append(tilt[i])
            if i+1==len(tilt):
                da_tilt.append((tilt[i]-tilt[i-1])/2.)
            else:
                da_tilt.append((tilt[i+1]-tilt[i])/2.)
            point_prob.extend(ones(nrot)*exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.extend(ones(nrot)*tilt[i])
            point_rot.extend(ones(nrot)*arange(0, 360, da_rot))

        elif abs(tilt[i]-180)<1e-10:
            rot_integral.append(exp(-U[i]/kT)*2*pi)
            angles.append(tilt[i])
            if i+1==len(tilt):
                da_tilt.append((tilt[i]-tilt[i-1])/2.)
            else:
                da_tilt.append((tilt[i+1]-tilt[i])/2.)
            point_prob.extend(ones(nrot)*exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.extend(ones(nrot)*tilt[i])
            point_rot.extend(ones(nrot)*arange(0, 360, da_rot))
            
        elif abs(tilt[i]-tilt[i-1])<1e-10 and i!=0:
            rot_integral[-1] += exp(-U[i]/kT)*da_rot*pi/180.
            point_prob.append(exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.append(tilt[i])
            point_rot.append(rot[i])
        else:
            rot_integral.append(exp(-U[i]/kT)*da_rot*pi/180.)
            angles.append(tilt[i])
            if i==0:
                ii=0
                da_tilt_aux = 0
                while da_tilt_aux==0:
                    da_tilt_aux = tilt[ii+1]-tilt[ii]
                    ii+=1
                da_tilt.append(da_tilt_aux)
            else:
                da_tilt.append(tilt[i]-tilt[i-1])
            point_prob.append(rot_integral[-1]*da_tilt[-1]*pi/180)
            point_tilt.append(tilt[i])
            point_rot.append(rot[i])

    print('%i tilt angles, from %f to %f in steps of %f'%(len(angles), angles[0], angles[-1], angles[1]-angles[0]))

    point_prob = array(point_prob)
    point_tilt = array(point_tilt)
    point_rot = array(point_rot)
    return rot_integral, da_tilt, angles, point_prob, point_tilt, point_rot

def tilt_integration(rot_integral, angles, angle_lim):

    tilt_integral = zeros(len(angle_lim)-1)
    for i in range(len(tilt_integral)):
        low = angle_lim[i]
        upp = angle_lim[i+1]
        #print '\nlow: %f, upp: %f'%(low,upp)
        low_limit=False

        for j in range(len(angles)-1):
            da_tilt = angles[j+1] - angles[j]
            if low<=angles[0]: 
                low = angles[0]+1e-10 
            if low>angles[j] and low <= angles[j+1]:
                #print 'Lower lim between %f and %f'%(angles[j],angles[j+1])
                if upp<angles[j+1]:
                    factor = (upp-low)/(angles[j+1]-angles[j])
                    tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                    #print 'Upper lim also between %f and %f'%(angles[j],angles[j+1])
                    break
                else:
                    factor = (angles[j+1]-low)/(angles[j+1]-angles[j])
                    tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                    low_limit = True


            elif low_limit==True and upp>angles[j+1]:
                tilt_integral[i] += da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                #print 'Full integration between %f and %f'%(angles[j],angles[j+1])

            elif upp <= angles[j+1]:
                factor = (upp-angles[j])/(angles[j+1]-angles[j])
                tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                #print 'Upper between %f and %f'%(angles[j],angles[j+1])
                break

    return tilt_integral



result_pos = loadtxt('3GHG_Graph_Results/E_int_3GHG-GRAPH_Conformacion_0-180_v2')
result_neg = loadtxt('3GHG_Graph_Results/E_int_3GHG-GRAPH_Conformacion_0-180_v2')


tilt_p = result_pos[:,0]
rot_p  = result_pos[:,1]
cal_p  = result_pos[:,2]

tilt_n = result_neg[:,0]
rot_n  = result_neg[:,1]
cal_n  = result_neg[:,2]

av_cal_n = average(cal_n)
cal_n -= av_cal_n

av_cal_p = average(cal_p)
cal_p -= av_cal_p

print('Negative pass')
rot_integral_n, da_tilt_n, angles_n, prob_point_n, til_point_n, rot_point_n = rotation_integration(cal_n, tilt_n, rot_n)
print('Integration goes from %f to %f'%(tilt_n[0], tilt_n[-1]))

print('Positive pass')
rot_integral_p, da_tilt_p, angles_p, prob_point_p, til_point_p, rot_point_p = rotation_integration(cal_p, tilt_p, rot_p)
print('Integration goes from %f to %f'%(tilt_p[0], tilt_p[-1]))


# Integration limits
#h = 0.005
h = 0.05
cos_th = arange(-1, 1+h, h)
cos_th[-1] = 1.
cos_th[0] = -1.


cos_th = flipud(cos_th) # Flip so that angles are in increasing order
angle_lim_cos = arccos(cos_th)*180/pi
angle_lim_alpha = arange(0,181,2)

tilt_integral_p_cos = tilt_integration(rot_integral_p, angles_p, angle_lim_cos) 
tilt_integral_p_alpha = tilt_integration(rot_integral_p, angles_p, angle_lim_alpha) 
tilt_integral_n_cos = tilt_integration(rot_integral_n, angles_n, angle_lim_cos) 
tilt_integral_n_alpha = tilt_integration(rot_integral_n, angles_n, angle_lim_alpha) 


cos_th_plot = zeros(len(cos_th)-1)
for i in range(len(cos_th_plot)):
    cos_th_plot[i] = (cos_th[i] + cos_th[i+1])/2

alpha_plot = zeros(len(angle_lim_alpha)-1)
for i in range(len(alpha_plot)):
    alpha_plot[i] = (angle_lim_alpha[i] + angle_lim_alpha[i+1])/2

Zn_cos = sum(tilt_integral_n_cos)
prob_n_cos = tilt_integral_n_cos/Zn_cos

Zn_alpha = sum(tilt_integral_n_alpha)
prob_n_alpha = tilt_integral_n_alpha/Zn_alpha

Zp_cos = sum(tilt_integral_p_cos)
prob_p_cos = tilt_integral_p_cos/Zp_cos

Zp_alpha = sum(tilt_integral_p_alpha)
prob_p_alpha = tilt_integral_p_alpha/Zp_alpha


cos_th_avg_n = sum(cos_th_plot*prob_n_cos)
alpha_avg_n = sum(alpha_plot*prob_n_alpha)

cos_th_avg_p = sum(cos_th_plot*prob_p_cos)
alpha_avg_p = sum(alpha_plot*prob_p_alpha)

print('cos(alpha):')
print('Average cos(th): \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(cos_th_avg_n, cos_th_avg_p))
print('Max probability: \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(max(prob_n_cos), max(prob_p_cos)))

print('alpha:')
print('Average angle  : \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(alpha_avg_n,alpha_avg_p))
print('Max probability: \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(max(prob_n_alpha), max(prob_p_alpha)))


font = {'family':'serif','size':10}

plt.rc('font',**font)

# Plot cos of angle
fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)
ax.plot(cos_th_plot, prob_n_cos, c='k', marker='', lw=0.8, ls='-', label=r'$3GHG$ - $Surf.$ $Graf.$ $478x475$')
#ax.plot(cos_th_plot, prob_p_cos, c='k', marker='', lw=0.8, ls=':')
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.set_ylabel('Probability')
ax.set_xlabel(r'cos($\alpha_{\rm tilt}$)')
ax.legend(fontsize=7, handlelength=2)
ax.text(0, max(prob_n_cos)/2,r'$\langle$cos($\alpha_{\rm tilt}$)$\rangle$=%1.3f'%(cos_th_avg_n),fontsize=6)
fig.subplots_adjust(left=0.25, bottom=0.21, right=0.96, top=0.95)
fig.savefig('3GHG_Graph_Results/probability_cos_3ghg-graph_478x475_v2.pdf')

# Plot angle
fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)
ax.plot(alpha_plot, prob_n_alpha, c='k', marker='', lw=0.8, ls='-', label=r'$3GHG$ - $Surf.$ $Graf.$ $478x475$')
#ax.plot(alpha_plot, prob_p_alpha, c='k', marker='', lw=0.8, ls=':', label=r'$ $')
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.set_ylabel('Probabilidad')
ax.set_xlabel(r'$\alpha_{\rm tilt}$')
ax.set_ylim(top=max(prob_p_alpha)*1.4)
ax.xaxis.set_ticks(arange(0,180+1,45))
ax.legend(loc='upper center', fontsize=7, handlelength=2).get_frame().set_lw(0.3)
ax.text(75, max(prob_n_cos)/2,r'$\langle \alpha_{\rm tilt}\rangle$=%1.1f$^{\circ}$'%(alpha_avg_n),fontsize=6)
fig.subplots_adjust(left=0.22, bottom=0.21, right=0.93, top=0.95)
fig.savefig('3GHG_Graph_Results/probability_alpha_3ghg-graph_478x475_v2.pdf')

# Scatter plot probability


fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)

til_point_p = reshape(til_point_p,(len(angles_p), int(len(til_point_p)/len(angles_p))))
rot_point_p = reshape(rot_point_p,(len(angles_p), int(len(rot_point_p)/len(angles_p))))
prob_point_p = reshape(prob_point_p,(len(angles_p), int(len(prob_point_p)/len(angles_p))))/Zp_alpha

rot_point_p_plot = zeros(shape(rot_point_p)+array([0,1]))
rot_point_p_plot[:,0:-1] = rot_point_p[:,:]
rot_point_p_plot[:,-1] = 360
da_rot = rot_point_p[0,1] - rot_point_p[0,0]
rot_point_p_plot -= da_rot/2

til_point_p_plot = zeros(shape(til_point_p)+array([0,1]))
til_point_p_plot[:,0:-1] = til_point_p[:,:]
til_point_p_plot[:,-1] = til_point_p_plot[:,-2]

#CS=ax.contourf(til_point_p, rot_point_p, prob_point_p)#, s=5, linewidths=0)
CS=ax.pcolor(til_point_p_plot, rot_point_p_plot, prob_point_p, cmap=plt.get_cmap('YlGnBu'))
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.xaxis.set_ticks(arange(0,180+1,45))
ax.set_ylabel(r'$\alpha_{\rm rot}$')
ax.set_xlabel(r'$\alpha_{\rm tilt}$')
ax.set_title(r'Probability 3GHG-Graphene', fontsize=8)
ax.set_ylim([0,350])
ax.set_xlim([0,180])
CB=fig.colorbar(CS)
#print (CB.ax.get_children())
CB.ax.get_children()[2].set_linewidth(0.5)
CB.ax.tick_params(labelsize=6)
fig.subplots_adjust(left=0.20, bottom=0.25, right=0.94, top=0.88)
fig.savefig('3GHG_Graph_Results/probability_graphene_config2_478x475_v2.pdf')


```

```python
import numpy
import sys
import os


################################################################################################################
#Archivos output proteina superficie interactuando
################################################################################################################

Output_interaccion = '1FZA-Graph_Results/1FZA-Interaccion_Phi-File_0-180'
file_aux= open('auxiliar.txt','w')
file_1 = open(Output_interaccion,'r')

E_solv_interaccion = []
h = []
a_tilt = []
a_rot = []
E_coul_interaccion = -96818963.231094

for line_1 in file_1:
    linea_1=line_1.split()
    if len(linea_1)==11:
        if linea_1[0]== '2.00':
            aa='{0} {1} {2} {3} \n'.format(linea_1[0],linea_1[2],linea_1[4],linea_1[6])
            file_aux.write(aa)
            
            
file_aux.close()

file_aux= open('auxiliar.txt','r')
X=numpy.loadtxt(file_aux)
Xc=numpy.copy(X)
E_solv_interaccion[:]=Xc[:,3]
h[:]=Xc[:,0]
a_tilt[:]=Xc[:,1]
a_rot[:]=Xc[:,2]
file_aux.close()


##############################################################################################################
#Funcion para leer archivo_output de PyGBe
##############################################################################################################

def scanOutput(filename):
    
    flag = 0 
    files = []
    file1=open(filename,'r')
    for line in file1:
        line = line.split()
        if len(line)>0:
            if line[0]=='Converged':
                iterations = int(line[2])
            if line[0]=='Total' and line[1]=='elements':
                N = int(line[-1])
            if line[0]=='Totals:':
                flag = 1 
            if line[0]=='Esolv' and flag==1:
                Esolv = float(line[2])
            if line[0]=='Esurf' and flag==1:
                Esurf = float(line[2])
            if line[0]=='Ecoul' and flag==1:
                Ecoul = float(line[2])
            if line[0]=='Time' and flag==1:
                Time = float(line[2])
            if line[0]=='Reading':
               files.append(line[-1])  

    return N, iterations, Esolv, Esurf, Ecoul, Time, files

##############################################################################################################
##############################################################################################################
#Proteina aislada
##############################################################################################################
N_prot_aislada = []
iterations_prot_aislada = []
Esolv_prot_aislada = []
Esurf_prot_aislada = []
Ecoul_prot_aislada = []
Time_prot_aislada = []

N_run, iterations_run, Esolv_run, Esurf_run, Ecoul_run, Time_run, files =scanOutput('1FZA-Graph_Results/1FZA_Aislada_out')

N_prot_aislada.append(N_run)
iterations_prot_aislada.append(iterations_run)
Esolv_prot_aislada.append(Esolv_run)
Esurf_prot_aislada.append(Esurf_run)
Ecoul_prot_aislada.append(Ecoul_run)
Time_prot_aislada.append(Time_run)

#####################################################################################################################
#####################################################################################################################
#Superficie Bidimensional de Grafeno
#####################################################################################################################
N_graph_aislada = []
iterations_graph_aislada = []
Esolv_graph_aislada = []
Esurf_graph_aislada = []
Ecoul_graph_aislada = []
Time_graph_aislada = []

N_run1, iterations_run1, Esolv_run1, Esurf_run1, Ecoul_run1, Time_run1, files1 =scanOutput('1FZA-Graph_Results/surf2d_252_Aislada_out')

N_graph_aislada.append(N_run1)
iterations_graph_aislada.append(iterations_run1)
Esolv_graph_aislada.append(Esolv_run1)
Esurf_graph_aislada.append(Esurf_run1)
Ecoul_graph_aislada.append(Ecoul_run1)
Time_graph_aislada.append(Time_run1)


#####################################################################################################################
#####################################################################################################################
#Calculo energia de interaccion y encontrar estado de minima Energia Libre de Interacción.

E_interaccion = numpy.zeros(len(h))

for i in range(len(h)):
    E_interaccion[i]= (E_solv_interaccion[i] + E_coul_interaccion) - (Esolv_prot_aislada[0] + Ecoul_prot_aislada[0] + Esolv_graph_aislada[0] + Ecoul_graph_aislada[0])
    
E_mas_prob=min(E_interaccion)
argumento_mas_prob=numpy.argmin(E_interaccion)
h_mas_prob = h[argumento_mas_prob]
a_tilt_mas_prob=a_tilt[argumento_mas_prob]
a_rot_mas_prob=a_rot[argumento_mas_prob]

print('Distancia proteina sensor =',h_mas_prob, ' Angstrom')
print('Alpha_tilt =', a_tilt_mas_prob, '°')
print('Alpha_rot =', a_rot_mas_prob, '°')
print('Energia de Interaccion =', E_mas_prob, ' [Kcal/mol]')

#Archivo_input_probability_plot_python
conformacion_file = open('1FZA-Graph_Results/E_int_1FZA-GRAPH_Conformacion_0-180_v2','w')
for i in range(len(h)):
    angulos_energia_interaccion='{0} {1} {2} \n'.format(a_tilt[i], a_rot[i], E_interaccion[i])
    conformacion_file.write(angulos_energia_interaccion)
conformacion_file.close()

os.system('rm -r auxiliar.txt')
os.system('rm -r auxiliar_2.txt')	


```

```python
from numpy import *
import matplotlib as mpl
import matplotlib.pyplot as plt

def rotation_integration(U, tilt, rot):

    i=0
    da_rot = 0
    while da_rot==0:
        da_rot = rot[i+1]-rot[i]
        i+=1
    print('da_rot = %f'%(da_rot))

    nrot = int(360/da_rot)
    point_prob = []
    point_tilt = []
    point_rot  = []

    kT = 2.479/4.184
    rot_integral = []
    angles = []
    da_tilt = []
    for i in range(len(tilt)):
        # Extreme cases: tilt = 0 and tilt = 180
        if abs(tilt[i])<1e-10:
            rot_integral.append(exp(-U[i]/kT)*2*pi)
            angles.append(tilt[i])
            if i+1==len(tilt):
                da_tilt.append((tilt[i]-tilt[i-1])/2.)
            else:
                da_tilt.append((tilt[i+1]-tilt[i])/2.)
            point_prob.extend(ones(nrot)*exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.extend(ones(nrot)*tilt[i])
            point_rot.extend(ones(nrot)*arange(0, 360, da_rot))

        elif abs(tilt[i]-180)<1e-10:
            rot_integral.append(exp(-U[i]/kT)*2*pi)
            angles.append(tilt[i])
            if i+1==len(tilt):
                da_tilt.append((tilt[i]-tilt[i-1])/2.)
            else:
                da_tilt.append((tilt[i+1]-tilt[i])/2.)
            point_prob.extend(ones(nrot)*exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.extend(ones(nrot)*tilt[i])
            point_rot.extend(ones(nrot)*arange(0, 360, da_rot))
            
        elif abs(tilt[i]-tilt[i-1])<1e-10 and i!=0:
            rot_integral[-1] += exp(-U[i]/kT)*da_rot*pi/180.
            point_prob.append(exp(-U[i]/kT)*da_tilt[-1]*da_rot*(pi/180)**2)
            point_tilt.append(tilt[i])
            point_rot.append(rot[i])
        else:
            rot_integral.append(exp(-U[i]/kT)*da_rot*pi/180.)
            angles.append(tilt[i])
            if i==0:
                ii=0
                da_tilt_aux = 0
                while da_tilt_aux==0:
                    da_tilt_aux = tilt[ii+1]-tilt[ii]
                    ii+=1
                da_tilt.append(da_tilt_aux)
            else:
                da_tilt.append(tilt[i]-tilt[i-1])
            point_prob.append(rot_integral[-1]*da_tilt[-1]*pi/180)
            point_tilt.append(tilt[i])
            point_rot.append(rot[i])

    print('%i tilt angles, from %f to %f in steps of %f'%(len(angles), angles[0], angles[-1], angles[1]-angles[0]))

    point_prob = array(point_prob)
    point_tilt = array(point_tilt)
    point_rot = array(point_rot)
    return rot_integral, da_tilt, angles, point_prob, point_tilt, point_rot

def tilt_integration(rot_integral, angles, angle_lim):

    tilt_integral = zeros(len(angle_lim)-1)
    for i in range(len(tilt_integral)):
        low = angle_lim[i]
        upp = angle_lim[i+1]
        #print '\nlow: %f, upp: %f'%(low,upp)
        low_limit=False

        for j in range(len(angles)-1):
            da_tilt = angles[j+1] - angles[j]
            if low<=angles[0]: 
                low = angles[0]+1e-10 
            if low>angles[j] and low <= angles[j+1]:
                #print 'Lower lim between %f and %f'%(angles[j],angles[j+1])
                if upp<angles[j+1]:
                    factor = (upp-low)/(angles[j+1]-angles[j])
                    tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                    #print 'Upper lim also between %f and %f'%(angles[j],angles[j+1])
                    break
                else:
                    factor = (angles[j+1]-low)/(angles[j+1]-angles[j])
                    tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                    low_limit = True


            elif low_limit==True and upp>angles[j+1]:
                tilt_integral[i] += da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                #print 'Full integration between %f and %f'%(angles[j],angles[j+1])

            elif upp <= angles[j+1]:
                factor = (upp-angles[j])/(angles[j+1]-angles[j])
                tilt_integral[i] += factor*da_tilt*pi/180.*(rot_integral[j]+rot_integral[j+1])/2
                #print 'Upper between %f and %f'%(angles[j],angles[j+1])
                break

    return tilt_integral



result_pos = loadtxt('1FZA-Graph_Results/E_int_1FZA-GRAPH_Conformacion_0-180_v2')
result_neg = loadtxt('1FZA-Graph_Results/E_int_1FZA-GRAPH_Conformacion_0-180_v2')


tilt_p = result_pos[:,0]
rot_p  = result_pos[:,1]
cal_p  = result_pos[:,2]
#jou_p  = result_pos[:,3]

tilt_n = result_neg[:,0]
rot_n  = result_neg[:,1]
cal_n  = result_neg[:,2]
#jou_n  = result_neg[:,3]

av_cal_n = average(cal_n)
cal_n -= av_cal_n

av_cal_p = average(cal_p)
cal_p -= av_cal_p

print('Negative pass')
rot_integral_n, da_tilt_n, angles_n, prob_point_n, til_point_n, rot_point_n = rotation_integration(cal_n, tilt_n, rot_n)
print('Integration goes from %f to %f'%(tilt_n[0], tilt_n[-1]))

print('Positive pass')
rot_integral_p, da_tilt_p, angles_p, prob_point_p, til_point_p, rot_point_p = rotation_integration(cal_p, tilt_p, rot_p)
print('Integration goes from %f to %f'%(tilt_p[0], tilt_p[-1]))

# Integration limits
h = 0.05
cos_th = arange(-1, 1+h, h)
cos_th[-1] = 1.
cos_th[0] = -1.


cos_th = flipud(cos_th) # Flip so that angles are in increasing order
angle_lim_cos = arccos(cos_th)*180/pi
angle_lim_alpha = arange(0,181,2)

tilt_integral_p_cos = tilt_integration(rot_integral_p, angles_p, angle_lim_cos) 
tilt_integral_p_alpha = tilt_integration(rot_integral_p, angles_p, angle_lim_alpha) 
tilt_integral_n_cos = tilt_integration(rot_integral_n, angles_n, angle_lim_cos) 
tilt_integral_n_alpha = tilt_integration(rot_integral_n, angles_n, angle_lim_alpha) 


cos_th_plot = zeros(len(cos_th)-1)
for i in range(len(cos_th_plot)):
    cos_th_plot[i] = (cos_th[i] + cos_th[i+1])/2

alpha_plot = zeros(len(angle_lim_alpha)-1)
for i in range(len(alpha_plot)):
    alpha_plot[i] = (angle_lim_alpha[i] + angle_lim_alpha[i+1])/2

Zn_cos = sum(tilt_integral_n_cos)
prob_n_cos = tilt_integral_n_cos/Zn_cos

Zn_alpha = sum(tilt_integral_n_alpha)
prob_n_alpha = tilt_integral_n_alpha/Zn_alpha

Zp_cos = sum(tilt_integral_p_cos)
prob_p_cos = tilt_integral_p_cos/Zp_cos

Zp_alpha = sum(tilt_integral_p_alpha)
prob_p_alpha = tilt_integral_p_alpha/Zp_alpha


cos_th_avg_n = sum(cos_th_plot*prob_n_cos)
alpha_avg_n = sum(alpha_plot*prob_n_alpha)

cos_th_avg_p = sum(cos_th_plot*prob_p_cos)
alpha_avg_p = sum(alpha_plot*prob_p_alpha)

print('cos(alpha):')
print('Average cos(th): \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(cos_th_avg_n, cos_th_avg_p))
print('Max probability: \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(max(prob_n_cos), max(prob_p_cos)))

print('alpha:')
print('Average angle  : \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(alpha_avg_n,alpha_avg_p))
print('Max probability: \n\tPositive surface charge: %f\n\tNegative surface charge: %f'%(max(prob_n_alpha), max(prob_p_alpha)))

font = {'family':'serif','size':10}

plt.rc('font',**font)

# Plot cos of angle
fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)
ax.plot(cos_th_plot, prob_n_cos, c='k', marker='', lw=0.8, ls='-', label=r'$1FZA$ - $Surf.$ $Graf.$ $252x252$')
#ax.plot(cos_th_plot, prob_p_cos, c='k', marker='', lw=0.8, ls=':')
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.set_ylabel('Probability')
ax.set_xlabel(r'cos($\alpha_{\rm tilt}$)')
ax.legend(fontsize=7, handlelength=2)
ax.text(0, max(prob_n_cos)/2,r'$\langle$cos($\alpha_{\rm tilt}$)$\rangle$=%1.3f'%(cos_th_avg_n),fontsize=6)
fig.subplots_adjust(left=0.25, bottom=0.21, right=0.96, top=0.95)
fig.savefig('1FZA-Graph_Results/probability_cos_1fza-graph_252x252_v2.pdf')

# Plot angle
fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)
ax.plot(alpha_plot, prob_n_alpha, c='k', marker='', lw=0.8, ls='-', label=r'$1FZA$ - $Surf.$ $Graf.$ $252x252$')
#ax.plot(alpha_plot, prob_p_alpha, c='k', marker='', lw=0.8, ls=':', label=r'$ $')
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.set_ylabel('Probabilidad')
ax.set_xlabel(r'$\alpha_{\rm tilt}$')
ax.set_ylim(top=max(prob_p_alpha)*1.4)
ax.xaxis.set_ticks(arange(0,180+1,45))
ax.legend(loc='upper center', fontsize=7, handlelength=2).get_frame().set_lw(0.3)
ax.text(75, max(prob_n_cos)/2,r'$\langle \alpha_{\rm tilt}\rangle$=%1.1f$^{\circ}$'%(alpha_avg_n),fontsize=6)
fig.subplots_adjust(left=0.22, bottom=0.21, right=0.93, top=0.95)
fig.savefig('1FZA-Graph_Results/probability_alpha_1fza-graph_252x252_v2.pdf')

# Scatter plot probability


fig = plt.figure(figsize=(3,2), dpi=80)
ax = fig.add_subplot(111)

til_point_p = reshape(til_point_p,(len(angles_p), int(len(til_point_p)/len(angles_p))))
rot_point_p = reshape(rot_point_p,(len(angles_p), int(len(rot_point_p)/len(angles_p))))
prob_point_p = reshape(prob_point_p,(len(angles_p), int(len(prob_point_p)/len(angles_p))))/Zp_alpha

rot_point_p_plot = zeros(shape(rot_point_p)+array([0,1]))
rot_point_p_plot[:,0:-1] = rot_point_p[:,:]
rot_point_p_plot[:,-1] = 360
da_rot = rot_point_p[0,1] - rot_point_p[0,0]
rot_point_p_plot -= da_rot/2

til_point_p_plot = zeros(shape(til_point_p)+array([0,1]))
til_point_p_plot[:,0:-1] = til_point_p[:,:]
til_point_p_plot[:,-1] = til_point_p_plot[:,-2]

#CS=ax.contourf(til_point_p, rot_point_p, prob_point_p)#, s=5, linewidths=0)
CS=ax.pcolor(til_point_p_plot, rot_point_p_plot, prob_point_p, cmap=plt.get_cmap('YlGnBu'))
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.xaxis.set_ticks(arange(0,180+1,45))
ax.set_ylabel(r'$\alpha_{\rm rot}$')
ax.set_xlabel(r'$\alpha_{\rm tilt}$')
ax.set_title(r'Probability 1FZA-Graphene', fontsize=8)
ax.set_ylim([0,350])
ax.set_xlim([0,180])
CB=fig.colorbar(CS)
#print (CB.ax.get_children())
CB.ax.get_children()[2].set_linewidth(0.5)
CB.ax.tick_params(labelsize=6)
fig.subplots_adjust(left=0.20, bottom=0.25, right=0.94, top=0.88)
fig.savefig('1FZA-Graph_Results/probability_graphene_config2_252x252_v2.pdf')
```
